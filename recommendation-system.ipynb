```python
import pandas as pd
import numpy as np
```

```python
df_movies = pd.read_csv("tmdb_5000_movies.csv")
df_credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
df_movies.head()
```

```python
print("credit Shape",df_credits.shape)
print("Movies Shape", df_movies.shape)
```

```python
df_credit_column_remaned = df_credits.rename(index=str, columns={"movie_id":"id"})
df_merged = df_movies.merge(df_credit_column_remaned, on="id")
print(df_merged)
```

```python
df_merged.head()
```

```python
df_cleaned = df_merged.drop(columns=["homepage","title_x", "title_y", "status", "production_countries"])
```

```python
df_cleaned.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3, max_features=None,
                    strip_accents='unicode', analyzer='word',
                    token_pattern=r'w{1,}',ngram_range=(1,3),
                    stop_words = 'english')
```

```python
tfv_mat = tfv.fit_transform(df_cleaned['overview'])
```

```python
df_cleaned.info()
```

```python
df_cleaned.dropna(subset=["overview"], inplace=True)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tfv_mat, tfv_mat)
```

```python
print(sig[0])
```

```python
indices = pd.Series(df_cleaned.index, index=df_cleaned['original_title']).drop_duplicates()
```

```python
def give_recommendation(title, sig=sig):
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    sig_scores = sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    
    return df_cleaned["original_title"].iloc[movie_indices]
```

```python
print(give_recommendation('Avatar'))
```
